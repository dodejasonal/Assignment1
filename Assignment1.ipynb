Q7- What is the difference between “==” and “is” operators in Python?

Ans- "==" is a Comparision Operator which compares two values and return a boolean value whereas "is" is an Identitiy operator which compares location of two object/variables.

Q8-What are logical operators in Python?

Ans- In Python, logical operators are used to perform logical operations on Boolean values (True or False). These operators allow you to combine multiple conditions or expressions in a more complex logical expression. Python provides three main logical operators:And , Or, Not.

Q9-What is type casting in Python?

Ans-Type casting is the process of changing the data type of value/object i.e fron string to integer etc.

Q10-What is the difference between implicit and explicit type casting?

Ans- Implicit type casting, also known as automatic type conversion, occurs when Python automatically converts a smaller data type to a larger data type to prevent data loss. This is done without the programmer's intervention whereas Explicit type casting, also known as manual type conversion, happens when the programmer explicitly converts one data type to another using functions like int(), float(), str(), etc.

Q11-What is the purpose of conditional statements in Python?

Ans- Conditional statements helps the programmer to code decisions based on some preconditions.

Q12- How does the elif statement work?

Ans- The elif statement in Python stands for "else if" and is used in conditional statements to check multiple conditions. It allows you to check for additional conditions after an initial if statement, without needing to write multiple if statements. It works as a chain of conditions where each condition is checked in order. If one of the conditions is True, the corresponding block of code is executed.

Q13-What is the difference between for and while loops?

Ans- While Loop is a loop that repeatedly execute a block of code until a condition is met whereas For Loop iterate over a swuence of elements.

Q14-Describe a scenario where a while loop is more suitable than a for loop.

Ans- A while loop is more suitable than a for loop when you don't know in advance how many times you need to repeat a block of code, and the loop should continue based on a condition that is evaluated during each iteration.

```python
#Q14- Write a program to reverse a string using a while loop?
string=input("Enter a string:")
reverse=""
i=len(string)-1
while i>=0:
  reverse+=string[i]
  i-=1
print("Reverse of the string is:",reverse)
```

```python
#Q15- Write a Python program to calculate the factorial of a number provided by the user using a while loop?
num=int(input("Enter a number:"))
fact=1
i=1
while i<=num:
  fact*=i
  i+=1
print("Factorial of",num,"is",fact)
```
